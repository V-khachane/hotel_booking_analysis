# Hypothesis Testing:
        1. Customers booking more than 6 months in advance are more likely to cancel.
        2. Weekday bookings have a higher average daily rate than weekend bookings.
        3. Guests with a higher number of special requests are more likely to cancel their bookings.

```python
# 1. Customers booking more than 6 months in advance are more likely to cancel.
     
     Null Hypothesis (H0): There is no difference in cancellation rates between customers who book more than 6 months in advance and those who do not.
     
     Alternative Hypothesis (H1): Customers who book more than 6 months in advance have a higher cancellation rate.
```

```python
# Use a chi-squared test to compare cancellation rates between the two groups.
from scipy import stats
```

```python
# Create a new column for booking in advance:
hotel_df1['booking_in_advance'] = hotel_df1['lead_time'] > 180                   # more than 6 months
```

```python
# Contingency table
contingency_table = pd.crosstab(hotel_df1['booking_in_advance'], hotel_df1['is_canceled'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
```

```python
# Printing the output:
print(f'Chi-squared: {chi2}, p-value: {p}')
```

```python
# 2. Weekday bookings have a higher average daily rate than weekend bookings.
     
    Null Hypothesis (H0): There is no difference in the average daily rate (ADR) between weekday and weekend bookings.

    Alternative Hypothesis (H1): Weekday bookings have a higher average daily rate than weekend bookings.
```

```python
# Use a t-test to compare the means of ADR for weekday and weekend bookings.
```

```python
# Create a new column for weekday vs. weekend
hotel_df1['is_weekend'] = hotel_df1['arrival_date_week_number'] % 2 == 0 
```

```python
# Separate the ADR values
weekday_adrs = hotel_df1[hotel_df1['is_weekend'] == False]['adr']
weekend_adrs = hotel_df1[hotel_df1['is_weekend'] == True]['adr']
```

```python
t_stat, p_value = stats.ttest_ind(weekday_adrs, weekend_adrs)

# Printing the Output:
print(f'T-statistic: {t_stat}, p-value: {p_value}')
```

```python
# 3. Guests with a higher number of special requests are more likely to cancel their bookings:

    Null Hypothesis (H0): There is no relationship between the number of special requests and the cancellation of bookings.

    Alternative Hypothesis (H1): Guests with a higher number of special requests are more likely to cancel their bookings.
```

```python
# Chi-squared test of independence:

# Create a contingency table
contingency_table = pd.crosstab(hotel_df1['total_of_special_requests'], hotel_df1['is_canceled'])

# Perform Chi-squared test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
```

```python
# Printing the Output:
print(f'Chi-squared: {chi2}, p-value: {p}')
```

```python
# Split the Data into Training and Test Sets:
from sklearn.model_selection import train_test_split
```

```python
features = hotel_df1[['lead_time', 'arrival_date_year', 'arrival_date_month', 'adults', 'children', 
                       'hotel', 'meal', 'market_segment', 'deposit_type', 'previous_cancellations', 
                       'previous_bookings_not_canceled']]
target = hotel_df1['is_canceled']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

# Train Two Different Models
    Logistic Regression
    Random Forest

```python
# Importing :

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report
```

```python
# Preprocessing
categorical_features = ['hotel', 'meal', 'market_segment', 'deposit_type']
numerical_features = ['lead_time', 'arrival_date_year', 'adults', 'children','previous_cancellations', 'previous_bookings_not_canceled']
```

```python
preprocessor = ColumnTransformer(transformers=[('num', 'passthrough', numerical_features),('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)])
```

```python
# Logistic Regression:

logistic_model = Pipeline(steps=[('preprocessor', preprocessor),('classifier', LogisticRegression(max_iter=400, solver='lbfgs', random_state=42))])
logistic_model.fit(X_train, y_train)
y_pred_logistic = logistic_model.predict(X_test)
```

```python
# Random Forest:

random_forest_model = Pipeline(steps=[('preprocessor', preprocessor),('classifier', RandomForestClassifier())])
random_forest_model.fit(X_train, y_train)
y_pred_forest = random_forest_model.predict(X_test)
```

```python
print("Logistic Regression Report:")
print(classification_report(y_test, y_pred_logistic))
```

```python
print("Random Forest Report:")
print(classification_report(y_test, y_pred_forest))
```
